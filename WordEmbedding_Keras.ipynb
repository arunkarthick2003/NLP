# Word embedding techniques using Embedding layer in Keras

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
sent
```

```python
#vocabulary size
voc_size=10000
```

One Hot Representation

```python
onehot_rep=[one_hot(words,voc_size)for words in sent]
print(onehot_rep)
```

**Word Embedding Representation**

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_len=8
#padding=pre addes 0 to the front
embedded_docs=pad_sequences(onehot_rep,padding='pre',maxlen=sent_len)
print(embedded_docs)
```

```python
dim=15
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
